```python
from sklearn.preprocessing import LabelEncoder
obj1 =LabelEncoder()
df["label"] = obj1.fit_transform(df["label"])
obj1.classes_
```

# **Scaling Data**

```python
from sklearn.preprocessing import MinMaxScaler
obj2=MinMaxScaler()
df[:]=obj2.fit_transform(df)
df.head()
```

# **Spliting Data**

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.iloc[:,:-1]
y=df['label']
```

```python
x_train1,x_test1,y_train,y_test=train_test_split(x,y)
```

```python
x_train2=x_train1[["meanfun","IQR","Q25"]]
x_test2=x_test1[["meanfun","IQR","Q25"]]
```

```python
from sklearn import svm
from sklearn.metrics import accuracy_score
```

# **Running Models and comparing results**

```python
list1=[]
kernel=['rbf','poly','linear']
for i in kernel:
        model=svm.SVC(kernel=i).fit(x_train2,y_train)
        y_pred=model.predict(x_test2)
        print(i, ' accuracy is: ', accuracy_score(y_test,y_pred))
        list1.append(accuracy_score(y_test,y_pred))
pos = list1.index(max(list1))
print("Highest accuracy is %s occurs at %s kernel." % (list1[pos], kernel[pos]))
```
